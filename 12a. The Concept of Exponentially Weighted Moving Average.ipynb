```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./datasets/DailyDelhiClimate.csv')
df.head()
```

```python
df_sub = df[['date', 'meantemp']]
df_sub.head()
```

```python
plt.scatter(df['date'], df['meantemp'], color = 'red')
```

```python
x1 = df['meantemp'].ewm(alpha = 0.9).mean()
x1
```

```python
df_sub['ewma'] = x1
df_sub.head()
```

```python
plt.scatter(df['date'], df['meantemp'], color = 'red')
plt.plot(df['date'], x1, color = 'black', label = 'alpha = 0.9')
plt.legend()
plt.show()
```
